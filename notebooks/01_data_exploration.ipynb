```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load processed data
prices = pd.read_csv("../data/processed/prices_usd.csv", index_col=0, parse_dates=True)
logrets = pd.read_csv("../data/processed/log_returns.csv", index_col=0, parse_dates=True)
```

```python
# Preview
display(prices.head())
```

```python
display(logrets.describe())
```

```python
# Missing values
sns.heatmap(prices.isna(), cbar=False)
plt.title("Missing values in prices")
plt.show()
```

```python
# Price series
prices.plot(figsize=(12,6))
plt.title("Asset Prices (USD)")
plt.show()
```

```python
# Correlation heatmap
sns.heatmap(logrets.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Log Return Correlations")
plt.show()
```

```python
# Distribution of returns
logrets.hist(bins=50, figsize=(12,6))
plt.suptitle("Log Return Distributions")
plt.show()
```
